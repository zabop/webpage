```python
import glob
```

```python
pngs=glob.glob('../zabop.github.io/OC/OC*')
```

```python
objIDs=[each[30:39] for each in pngs]
```

```python
classedas=["BLANK" for each in objIDs]
```

```python
intro='''
<!DOCTYPE html>
<html>

<head>
<style>
img {
    display: block;
    margin-left: auto;
    margin-right: auto;
}
h2,p {
    text-align: center;
}

hr {
width: 80%;
height: 8px;
margin-left: auto;
margin-right: auto;
background-color:#FF0066;
color:#FF0066;
border: 0 none;
}


</style>
</head>

<body align="center">
<hr>
'''
```

```python
def middle(index):

    midtext="""
<h1>ObjectID: """+str(objIDs[index])+"""</h1>
<h2>Classified as: """+classedas[index]+'''</h2>
<img src="'''+pngs[index]+'''">

<button onclick="myFunction_'''+str(objIDs[index])+'''()">Yes (correct classification)</button>
<button onclick="myFunction2_'''+str(objIDs[index])+'''()">No (incorrect classification)</button>
<button onclick="myFunction3_'''+str(objIDs[index])+'''()">Uncertain classification</button>
<input type="text" id="myText_'''+str(objIDs[index])+'''" value="Uncertain:">
<p id="demo_'''+str(objIDs[index])+'''"></p>
<hr>

<script>
function myFunction_'''+str(objIDs[index])+'''() {
  document.getElementById("demo_'''+str(objIDs[index])+'''").innerHTML = "Yes";
}
</script>

<script>
function myFunction2_'''+str(objIDs[index])+'''() {
  document.getElementById("demo_'''+str(objIDs[index])+'''").innerHTML = "No";
}
</script>

<script>
function myFunction3_'''+str(objIDs[index])+'''() {
  document.getElementById("demo_'''+str(objIDs[index])+'''").innerHTML = "Uncertain";
  var x = document.getElementById("myText_'''+str(objIDs[index])+'''").value;
  document.getElementById("demo_'''+str(objIDs[index])+'''").innerHTML = x;
}
</script>
'''
    return midtext
```

```python
outro='''
</body>
</html>
'''
```

```python
fout=open('htmlsource3.html','w')
fout.write(intro)
for INDEX, EACH in enumerate(pngs): fout.write(middle(INDEX))
fout.write(outro)
fout.close()
```
